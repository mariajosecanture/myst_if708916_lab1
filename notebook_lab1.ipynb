La inversión activa consiste en invertir en acciones y bonos con el objetivo de conseguir mejores retornos que la media del mercado. La inversión activa es batir el mercado. Un gastor de fondos de inversión activa elige las acciones que cree que van a hacerlo mejor que el mercado para de este modo conseguir rentabilidades más altas que la media.

El objetivo de la inversión pasiva es seguir un índice específico, se centra en intentar igualar el retorno del mercado lo más cercanamente posible.
